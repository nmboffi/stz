```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('output.txt')
```

```python
# Time
t = data[:, 0]

# x values
x = data[:, 1]

# velocity values
v = data[:, 4]

# sigma_12 (assumed shear wave simulation)
sig = data[:, 5]

# Analytical velocity solution
ana_v = data[:, 6]

# Analytical sigma solution
ana_sig = data[:, 7]

# Error in sigma
sig_err = data[:, -1]

# Error in velocity
u_err = data[:, -2]
```

```python
# Take a look at the maximum error that shows up 
print max(sig_err)
print max(u_err)
print max(v)
```

```python
# Look at the error as a function of time at a given x value
plt.plot(t[0::100], u_err[30::100])
```

```python
# Plot the velocity solutions as a function of t for fixed x
plt.plot(t[0::100], v[50::100])
plt.plot(t[0::100], ana_v[50::100])
```

```python
# Plot the sigma solutions as a function of time at a fixed x
plt.plot(t[0::100], sig[50::100])
plt.plot(t[0::100], ana_sig[50::100])
```
